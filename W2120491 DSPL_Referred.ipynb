```python
#Building the regression model
X = merged[[
        'GHG_Total',
        'Nitrous_Oxide',
        'Methane',
        'CO2',
        'Annual_GHG_Emissions',
        'Fossil_Fuels',
        'Land_Use']]
y = merged['Temperature_Anomaly']
```

```python
train = merged[
    merged['Year'] <= 1990]

test = merged[
    merged['Year'] > 1990]
feature_cols = [
    'GHG_Total',
    'Nitrous_Oxide',
    'Methane',
    'CO2',
    'Annual_GHG_Emissions',
    'Fossil_Fuels',
    'Land_Use']

X_train = train[feature_cols]
y_train = train['Temperature_Anomaly']

X_test = test[feature_cols]
y_test = test['Temperature_Anomaly']
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(
    X_train,
    y_train)
```

```python
predictions = model.predict(
    X_test
)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
import numpy as np
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

mae = mean_absolute_error(
    y_test,
    predictions
)

rmse = np.sqrt(
    mean_squared_error(
        y_test,
        predictions
    )
)

r2 = r2_score(
    y_test,
    predictions
)

print("MAE:", mae)
print("RMSE:", rmse)
print("R2:", r2)
```

```python
plt.figure(figsize=(12,6))

plt.plot(
    test['Year'],
    y_test,
    label='Actual')

plt.plot(
    test['Year'],
    predictions,
    label='Predicted')

plt.legend()
plt.title(
    'Actual vs Predicted Temperature Anomaly')
plt.show()
```

```python
# Re-load df3_all and prepare top10 for this specific plot
df3_all = pd.read_csv("/content/drive/MyDrive/DSPL /Dataset/ghg-emissions-by-world-region.csv")
latest_year = df3_all['Year'].max()
top10_df3 = df3_all[df3_all['Year'] == latest_year]
top10_df3 = top10_df3[top10_df3['Entity'] != 'World']
top10 = top10_df3.nlargest(10, 'Annual greenhouse gas emissions in CO? equivalents')

# Rename the column to 'GHG_Emissions' for consistency with the rest of the plotting code
top10 = top10.rename(columns={'Annual greenhouse gas emissions in CO? equivalents': 'GHG_Emissions'})

top10 = top10.sort_values(
    by='GHG_Emissions',
    ascending=True)
plt.figure(figsize=(12,7))
bars = plt.barh(
    top10['Entity'],
    top10['GHG_Emissions'])
plt.title(
    f'Top 10 Greenhouse Gas Emitting Countries ({latest_year})',
    fontsize=16)
plt.xlabel('GHG Emissions (CO₂ Equivalents)')
plt.ylabel('Country')
for bar in bars:
    width = bar.get_width()
    plt.text(
        width,
        bar.get_y() + bar.get_height()/2,
        f'{width/1e9:.1f}B',
        va='center' )
plt.tight_layout()
plt.show()
```

```python
df3_all = pd.read_csv("/content/drive/MyDrive/DSPL /Dataset/ghg-emissions-by-world-region.csv")
earliest_year = df3_all['Year'].min()
latest_year = df3_all['Year'].max()

earliest_data_ghg = df3_all[df3_all['Year'] == earliest_year]
latest_data_ghg = df3_all[df3_all['Year'] == latest_year]

# Merge dataframes on 'Entity' to calculate growth
growth_calculation_ghg = pd.merge(
    earliest_data_ghg,
    latest_data_ghg,
    on='Entity',
    suffixes=('_earliest', '_latest'))

growth_calculation_ghg = growth_calculation_ghg[~growth_calculation_ghg['Entity'].isin(['World', 'Asia', 'Europe', 'North America', 'Africa', 'South America', 'Oceania', 'European Union', 'International transport', 'High-income countries', 'Upper-middle-income countries', 'Lower-middle-income countries', 'OECD (Jones et al. 2023)'])]
growth_calculation_ghg['Annual greenhouse gas emissions in CO? equivalents_earliest'] = growth_calculation_ghg['Annual greenhouse gas emissions in CO? equivalents_earliest'].replace(0, pd.NA)
# Calculate percentage growth
growth_calculation_ghg['Growth_Percentage'] = (
    (growth_calculation_ghg['Annual greenhouse gas emissions in CO? equivalents_latest'] -
     growth_calculation_ghg['Annual greenhouse gas emissions in CO? equivalents_earliest']) /
    growth_calculation_ghg['Annual greenhouse gas emissions in CO? equivalents_earliest']
) * 100
# Create top10_growth and sort by growth percentage, dropping NaN values
top10_growth = growth_calculation_ghg[
    ['Entity', 'Growth_Percentage']
].rename(columns={'Entity': 'Country'}) \
.dropna(subset=['Growth_Percentage']) \
.sort_values(by='Growth_Percentage', ascending=False)

top10_growth = top10_growth.head(10)

plt.figure(figsize=(12,7))

bars = plt.barh(
    top10_growth['Country'],
    top10_growth['Growth_Percentage'])
plt.title(
    'Top 10 Fastest Growing Greenhouse Gas Emitting Countries',
    fontsize=16)
plt.xlabel('Growth in Emissions (%)')
plt.ylabel('Country')
for bar in bars:
    width = bar.get_width()

    plt.text(
        width,
        bar.get_y() + bar.get_height()/2,
        f'{width:.1f}%',
        va='center'
    )
plt.tight_layout()
plt.show()
```
